# Notebook 3:  Redes de Kohonen y SOM (Self organized maps)
En este notebook vamos a implementar el algoritmo de clustering SOM en R. El objetivo es entender como funciona este algoritmo y  aprender a utilizarlo con un ejemplo simple. Recuerde instalar previamente el paquete kohonen en R.



```r
library(kohonen)
set.seed(7)

#ruta datos
#lee datos del archivo de texto como una matriz. La primera #columna se utiliza como vector de nombres de las filas
datos <-read.table("milk1.dat",row.names=1,dec=".",header=TRUE)
datos <- as.data.frame(datos)
summary(datos)
```

```r
#create SOM grid
sommap <- som(scale(datos), grid = somgrid(2, 2, "hexagonal"))
plot(sommap)

colors <- function(n, alpha = 1) {
    rev(heat.colors(n, alpha))
}

plot(sommap, type = "counts", palette.name = colors, heatkey = TRUE)

plot(sommap, type = "mapping", pchs = 20, main = "Mapping Type SOM")

plot(sommap, type = "dist.neighbours", palette.name = terrain.colors)
```

Ejercicio 1: 
a) Explique como funciona el método som, describa parámetros de entrada y salida y los gráficos obtenidos.



#### a) El método som consiste básicamente en construir mapas autoorganizados de ciertos datos entregados.
Parámetros:
Entrada:
data: Matriz de datos entrenamiento.
xdim: largo de las grillas del mapa.
ydim: alto de las grillas del mapa
zim: Profundidad de la cuadrícula, hace que la cuadrícula sea 3D
rlen: Número de veces para recorrer los datos de entrenamiento para cada MST
alphaA: Comenzar y finalizar la tasa de aprendizaje
radiusA: Comienza y termina radio.
alphaB: Inicio y finalización de la tasa de aprendizaje para el segundo radio
radiusB: Radio de inicio y fin (asegúrese de que sea más grande que el radiusA)
init: Inicializar centros de clúster de forma no aleatoria
initf: Use la función de inicialización dada si init == T (predeterminado: Initialize_PCA)
distf: función de distancia.
codes: Centros de clúster para comenzar
importances: array de valores numéricos.
nhbr.method: Manera de computarizar la distancia de las grillas.
negRadius:Forma fácil de establecer el radio B como un múltiplo del radio predeterminado (use un valor más bajo para dimensiones más altas)
negAlpha: Lo mismo, solo que para alphaB
noMapping: Si es verdadero, no produce mapeo (por defecto Falso). Útil para uso en línea / streaming.

Salida:
Un mapa, que es una lista que contiene todos los parámetros y resultados

b) Modifique parámetros de entrada. Interprete y comente los resultados obtenidos. 

#### Modificando las dimensiones de la cuadrícula, podemos ver la cantidad de conjuntos que se generan, funcionando como una matriz, en este caso la muestra fué separada en nueve conjuntos. De aquí podemos apreciar que hay conjuntos con propiedades muy similares, tal como los de la posición [1,3], el [2,3] y el [3,3].

```r
sommap <- som(scale(datos), grid = somgrid(3, 3, "hexagonal"))
plot(sommap)
```

```r
mat <- matrix(sommap$codes[[1]], ncol=4,nrow=4,byrow=FALSE)

## use hierarchical clustering to cluster the codebook vectors
groups<-3
som.hc <- cutree(hclust(dist(mat)), groups)

#plot
plot(sommap, type="codes", bgcol=rainbow(groups)[som.hc])

#cluster boundaries
add.cluster.boundaries(sommap, som.hc)
```
